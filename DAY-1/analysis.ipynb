```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

# Make sure hotel_bookings.csv is in the same folder
df = pd.read_csv("hotel_bookings.csv")

print("="*60)
print("HOTEL BOOKING DEMAND DATASET")
print("="*60)

# ----------------------------------------------------------
# Dataset Overview
# ----------------------------------------------------------

print("\nFirst 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())

# ----------------------------------------------------------
# Dataset Shape
# ----------------------------------------------------------

print("\nDataset Shape")
print(df.shape)

print(f"\nRows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

# ----------------------------------------------------------
# Column Names
# ----------------------------------------------------------

print("\nColumn Names")
print(df.columns.tolist())

# ----------------------------------------------------------
# Data Types
# ----------------------------------------------------------

print("\nData Types")
print(df.dtypes)

# ----------------------------------------------------------
# Missing Values
# ----------------------------------------------------------

print("\nMissing Values")
print(df.isnull().sum())

print("\nTotal Missing Values")
print(df.isnull().sum().sum())

# ----------------------------------------------------------
# Duplicate Rows
# ----------------------------------------------------------

print("\nDuplicate Rows")
print(df.duplicated().sum())

# ----------------------------------------------------------
# Summary Statistics (Numerical)
# ----------------------------------------------------------

print("\nSummary Statistics (Numerical)")
print(df.describe())

# ----------------------------------------------------------
# Summary Statistics (Categorical)
# ----------------------------------------------------------

print("\nSummary Statistics (Categorical)")
print(df.describe(include='object'))

# ----------------------------------------------------------
# Unique Values
# ----------------------------------------------------------

print("\nUnique Values in Each Column")
print(df.nunique())

# ----------------------------------------------------------
# Target Variable
# ----------------------------------------------------------

print("\nTarget Variable : is_canceled")

print("\nBooking Cancellation Count")
print(df["is_canceled"].value_counts())

print("\nBooking Cancellation Percentage")
print((df["is_canceled"].value_counts(normalize=True)*100).round(2))

# ----------------------------------------------------------
# Correlation
# ----------------------------------------------------------

print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))

# ----------------------------------------------------------
# Basic Visualizations
# ----------------------------------------------------------

plt.figure(figsize=(6,4))
df["is_canceled"].value_counts().plot(kind="bar")
plt.title("Booking Cancellation")
plt.xlabel("Cancelled")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8,5))
df["hotel"].value_counts().plot(kind="bar")
plt.title("Hotel Type Distribution")
plt.xlabel("Hotel")
plt.ylabel("Bookings")
plt.show()

plt.figure(figsize=(8,5))
df["arrival_date_month"].value_counts().sort_index().plot(kind="bar")
plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Bookings")
plt.xticks(rotation=45)
plt.show()

# ----------------------------------------------------------
# Business Problem
# ----------------------------------------------------------

print("\n" + "="*60)
print("BUSINESS PROBLEM")
print("="*60)

print("""
Hotels lose significant revenue due to booking cancellations.

By predicting whether a booking is likely to be cancelled,
the hotel can:

• Reduce revenue loss
• Improve room allocation
• Manage overbooking
• Offer targeted discounts
• Improve customer satisfaction
""")

# ----------------------------------------------------------
# ML Problem Framing
# ----------------------------------------------------------

print("="*60)
print("MACHINE LEARNING PROBLEM")
print("="*60)

print("""
Problem Type : Classification

Reason:
The target variable 'is_canceled' has two classes:

0 -> Booking Not Cancelled
1 -> Booking Cancelled

Hence, this is a Binary Classification problem.
""")

# ----------------------------------------------------------
# Target Variable
# ----------------------------------------------------------

print("="*60)
print("TARGET VARIABLE")
print("="*60)

print("Target Variable : is_canceled")

# ----------------------------------------------------------
# Key Features
# ----------------------------------------------------------

print("="*60)
print("KEY FEATURES")
print("="*60)

key_features = [
    "hotel",
    "lead_time",
    "arrival_date_month",
    "stays_in_weekend_nights",
    "stays_in_week_nights",
    "adults",
    "children",
    "babies",
    "meal",
    "country",
    "market_segment",
    "distribution_channel",
    "previous_cancellations",
    "previous_bookings_not_canceled",
    "reserved_room_type",
    "assigned_room_type",
    "booking_changes",
    "deposit_type",
    "customer_type",
    "adr",
    "total_of_special_requests"
]

for feature in key_features:
    print("-", feature)

# ----------------------------------------------------------
# Three Key Observations
# ----------------------------------------------------------

print("\n" + "="*60)
print("THREE KEY OBSERVATIONS")
print("="*60)

print("""
Observation 1:
The dataset contains both numerical and categorical features,
making it suitable for multiple machine learning models.

Observation 2:
Columns such as company, agent, and country contain missing
values that need preprocessing before model training.

Observation 3:
Historical customer behavior like previous cancellations,
booking changes, lead time, and special requests are likely
to be strong predictors of booking cancellation.
""")

# ----------------------------------------------------------
# Conclusion
# ----------------------------------------------------------

print("="*60)
print("CONCLUSION")
print("="*60)

print("""
Dataset : Hotel Booking Demand

Business Problem:
Predict booking cancellations.

Machine Learning Type:
Binary Classification

Target Variable:
is_canceled

This dataset is well suited for predictive analytics and can
help hotels improve revenue management and operational efficiency.
""")
```
